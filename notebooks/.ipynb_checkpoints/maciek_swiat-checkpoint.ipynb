```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.optimize import curve_fit
```

```python
sns.set_style('darkgrid')
sns.set(rc={'figure.figsize':(15,8)})
```

```python
df= pd.read_csv('./../data/running_data.csv')
df2 = pd.read_csv('./../data/world_records.csv')
df2['Distance'] = df2['Distance']/1000
```

```python
m_records = df.loc[df['Type'] == 'Records']
m_records = m_records[['Moving Time', 'Distance']]
m_time = m_records['Moving Time']
m_distance = m_records['Distance']
```

```python
w_records = df2.loc[(m_distance.min() <= df2['Distance']) & (df2['Distance'] <= m_distance.max())]
w_time = w_records['Time']
w_distance = w_records['Distance']
```

------
## Porównanie rekordów

```python
f = lambda x,a,b,c: a*x**2+b*x+c
fit = curve_fit(f,m_distance,m_time)
fit2 = curve_fit(f,w_distance,w_time)
a_m, b_m, c_m= fit[0]
a_w, b_w, c_w = fit2[0]
```

```python
plt.scatter(x=m_distance, y=m_time, color='skyblue', label='Rekory życiowe')
plt.plot(m_distance, [f(x, a_m, b_m, c_m) for x in m_distance], color='skyblue')
plt.scatter(x=w_distance, y=w_time, color='darkblue', label='Rekordy świata')
plt.plot(w_distance, [f(x, a_w, b_w, c_w) for x in w_distance], color='darkblue')
plt.legend()
plt.xlabel('dystans [km]',fontsize=14,fontweight='bold')
plt.ylabel('czas [s]',fontsize=14,fontweight='bold')
plt.title('Porównanie rekordów na poszczególnych dystansach',fontweight='bold',fontsize=18)
plt.show()
```

-----
## Tempo

```python
tempo_m = m_time/m_distance
tempo_w = w_time/w_distance
tempo_min_m = (m_time/60)/m_distance
tempo_min_w = (w_time/60)/w_distance
```

```python
plt.scatter(x=m_distance, y=tempo_m, color='skyblue', label='Rekory życiowe')
plt.scatter(x=w_distance, y=tempo_w, color='darkblue', label='Rekordy świata')
plt.legend()
plt.xlabel('dystans [km]',fontsize=14,fontweight='bold')
plt.ylabel(r'tempo $\left[\frac{s}{km}\right]$',fontsize=14,fontweight='bold')
plt.title('Porównanie tempa na poszczególnych dystansach',fontweight='bold',fontsize=18)
plt.show()
```

```python
plt.scatter(x=m_time, y=tempo_m, color='skyblue', label='Rekory życiowe')
plt.scatter(x=w_time, y=tempo_w, color='darkblue', label='Rekordy świata')
plt.legend()
plt.xlabel('czas [s]',fontsize=14,fontweight='bold')
plt.ylabel(r'tempo $\left[\frac{s}{km}\right]$',fontsize=14,fontweight='bold')
plt.title('Porównanie tempa w zależności od czasu',fontweight='bold',fontsize=18)
plt.show()
```

```python
plt.scatter(x=m_distance, y=tempo_min_m, color='skyblue', label='Rekory życiowe')
plt.scatter(x=w_distance, y=tempo_min_w, color='darkblue', label='Rekordy świata')
plt.legend()
plt.xlabel('dystans [km]',fontsize=14,fontweight='bold')
plt.ylabel(r'tempo $\left[\frac{min}{km}\right]$',fontsize=14,fontweight='bold')
plt.title('Porównanie tempa na poszczególnych dystansach',fontweight='bold',fontsize=18)
plt.show()
```

```python
plt.scatter(x=m_time/60, y=tempo_min_m, color='skyblue', label='Rekory życiowe')
plt.scatter(x=w_time/60, y=tempo_min_w, color='darkblue', label='Rekordy świata')
plt.legend()
plt.xlabel('czas [min]',fontsize=14,fontweight='bold')
plt.ylabel(r'tempo $\left[\frac{min}{km}\right]$',fontsize=14,fontweight='bold')
plt.title('Porównanie tempa w zależności od czasu',fontweight='bold',fontsize=18)
plt.show()
```
